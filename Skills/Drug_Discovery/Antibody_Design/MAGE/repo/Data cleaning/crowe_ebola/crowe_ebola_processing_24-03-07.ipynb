```python
elisa_cols = ['EBOV', 'BDBV' , 'SUDV']
```

```python
binding_df = supp_table[elisa_cols]
binding_df['sGP'] = supp_table[['sGP ELISA1', 'sGP ELISA2']].mean(axis=1)

binding_df = binding_df.apply(lambda x: x >= 3)

binding_df['EBOV'] = binding_df['EBOV'].map({True:ebov_gp})
binding_df['BDBV'] = binding_df['BDBV'].map({True:bdbv_gp})
binding_df['SUDV'] = binding_df['SUDV'].map({True:sudv_gp})
binding_df['sGP'] = binding_df['sGP'].map({True:sGP})
```

```python
binding_df['clone'] = supp_table['Cluster Number']
binding_df = binding_df.melt(ignore_index=False, value_vars = ['EBOV', 'BDBV', 'SUDV', 'sGP'], id_vars='clone')
```

```python
binding_df = binding_df.dropna()
binding_df = binding_df.drop_duplicates(subset=['clone', 'variable'])
```

```python
binding_df['VH_AA'] = binding_df.index
binding_df['VH_AA'] = binding_df['VH_AA'].map(dict(zip(supp_table.index, supp_table['heavy_aa'])))

binding_df['VL_AA'] = binding_df.index
binding_df['VL_AA'] = binding_df['VL_AA'].map(dict(zip(supp_table.index, supp_table['light_aa'])))
```

```python
binding_df = binding_df.rename(columns = {'variable':'antigen_name', 'value':'antigen_seq'})
binding_df = binding_df.rename(columns = {})
```
